```python
# Import Libraries
import os
import csv
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from tqdm.auto import tqdm
import matplotlib.pyplot as plt
from torchmetrics.image.fid import FrechetInceptionDistance

# Generator Model
class Generator(nn.Module):
    def __init__(self, noise_dim, condition_dim, img_channels):
        super(Generator, self).__init__()
        self.fc = nn.Sequential(
            nn.Linear(noise_dim + condition_dim, 2 * 2 * 256, bias=False),
            nn.BatchNorm1d(2 * 2 * 256),
            nn.ReLU(inplace=True)
        )

        self.conv_blocks = nn.Sequential(
            nn.ConvTranspose2d(256, 128, kernel_size=5, stride=2, padding=2, output_padding=1, bias=False),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.ConvTranspose2d(128, 128, kernel_size=5, stride=2, padding=2, output_padding=1, bias=False),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.ConvTranspose2d(128, 128, kernel_size=5, stride=2, padding=2, output_padding=1, bias=False),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.ConvTranspose2d(128, img_channels, kernel_size=5, stride=2, padding=2, output_padding=1, bias=False),
            nn.Tanh()
        )

    def forward(self, noise, condition):
        merged = torch.cat((noise, condition), dim=1)
        out = self.fc(merged)
        out = out.view(out.size(0), 256, 2, 2)
        out = self.conv_blocks(out)
        return out

# Discriminator Model
class Discriminator(nn.Module):
    def __init__(self, img_channels, condition_dim):
        super(Discriminator, self).__init__()
        self.img_channels = img_channels  # Fix: Ensure this attribute is properly defined
        self.embedding = nn.Linear(condition_dim, img_channels, bias=False)

        self.conv_blocks = nn.Sequential(
            nn.Conv2d(img_channels * 2, 64, kernel_size=5, stride=2, padding=2),
            nn.LayerNorm([64, 16, 16]),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(64, 128, kernel_size=5, stride=2, padding=2),
            nn.LayerNorm([128, 8, 8]),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(128, 256, kernel_size=5, stride=2, padding=2),
            nn.LayerNorm([256, 4, 4]),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(256, 1, kernel_size=4, stride=1, bias=False),
        )

    def forward(self, img, condition):
        batch_size = img.size(0)
        cond_embed = self.embedding(condition).view(batch_size, self.img_channels, 1, 1)
        cond_embed = cond_embed.expand(-1, -1, img.size(2), img.size(3))
        merged_input = torch.cat((img, cond_embed), dim=1)
        out = self.conv_blocks(merged_input)
        return out

    
# Utility Functions
def get_noise(n_samples, z_dim, device='cpu'):
    return torch.randn(n_samples, z_dim, device=device)

def gradient_penalty(critic, real, fake, condition, device, lambda_gp=10):
    batch_size = real.size(0)
    epsilon = torch.rand(batch_size, 1, 1, 1, device=device, requires_grad=True)
    interpolated = epsilon * real + (1 - epsilon) * fake
    interpolated = interpolated.to(device)
    interpolated.requires_grad_(True)
    interpolated_score = critic(interpolated, condition)
    gradients = torch.autograd.grad(
        outputs=interpolated_score,
        inputs=interpolated,
        grad_outputs=torch.ones_like(interpolated_score, device=device),
        create_graph=True,
        retain_graph=True,
    )[0]
    gradients = gradients.view(batch_size, -1)
    gradient_norm = gradients.norm(2, dim=1)
    penalty = lambda_gp * ((gradient_norm - 1) ** 2).mean()
    return penalty

def show_real_and_fake_images(real_images, fake_images, title="Real vs. Fake Images"):
    """
    Display 6 real images alongside 6 corresponding fake images.
    """
    real_images = (real_images[:6] + 1) / 2  # Scale back to [0, 1] for visualization
    fake_images = (fake_images[:6] + 1) / 2  # Scale back to [0, 1] for visualization
    
    fig, axes = plt.subplots(2, 6, figsize=(12, 4))
    for i in range(6):
        axes[0, i].imshow(real_images[i].detach().cpu().permute(1, 2, 0).squeeze(), cmap='gray')
        axes[0, i].axis('off')
        axes[1, i].imshow(fake_images[i].detach().cpu().permute(1, 2, 0).squeeze(), cmap='gray')
        axes[1, i].axis('off')
    plt.suptitle(title)
    plt.show()
    
def plot_losses(gen_losses, crit_losses, title="Losses"):
    """
    Plot generator and critic losses.
    """
    plt.figure(figsize=(10, 5))
    plt.plot(gen_losses, label="Generator Loss")
    plt.plot(crit_losses, label="Critic Loss")
    plt.title(title)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()
    
def denormalize_to_uint8(images):
    """
    Convert normalized images ([-1, 1]) to uint8 format ([0, 255]).
    """
    return ((images + 1) * 127.5).clamp(0, 255).to(torch.uint8)
    
# Hyperparameters
BATCH_SIZE = 128
IMAGE_SIZE = 32
IMAGE_CHANNELS = 3
N_IT = 3
LAMBDA = 5
ALPHA = 3e-3
B1 = 0.5
B2 = 0.9
EPOCHS = 500
NOISE_DIM = 100
CONDITION_DIM = 10
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# Data Preparation
transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.5], [0.5])
])
dataset = datasets.CIFAR10(root="./data", transform=transform, download=True)
dataloader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)

# Models
generator = Generator(NOISE_DIM, CONDITION_DIM, IMAGE_CHANNELS).to(device)
discriminator = Discriminator(IMAGE_CHANNELS, CONDITION_DIM).to(device)

# Optimizers
gen_opt = optim.Adam(generator.parameters(), lr=5e-5, betas=(B1, B2))
crit_opt = optim.Adam(discriminator.parameters(), lr=2e-4, betas=(B1, B2))

# FID Metric
fid_metric = FrechetInceptionDistance(feature=64).to(device)

# Directories
os.makedirs("logs", exist_ok=True)
os.makedirs("models", exist_ok=True)

# Log file for losses and FID scores
log_file = "logs/training_log.csv"
with open(log_file, mode="w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Epoch", "Generator Loss", "Critic Loss", "FID Score"])

# Training Loop
generator_losses = []
critic_losses = []
fid_scores = []
iteration = 0  # Track iterations

for epoch in range(EPOCHS):
    epoch_gen_loss = 0
    epoch_crit_loss = 0
    for real_images, labels in tqdm(dataloader, desc=f"Epoch {epoch+1}/{EPOCHS}"):
        real_images = real_images.to(device)
        conditions = nn.functional.one_hot(labels, num_classes=CONDITION_DIM).float().to(device)

        # Train Critic
        for _ in range(N_IT):
            noise = get_noise(real_images.size(0), NOISE_DIM, device=device)
            fake_images = generator(noise, conditions).detach()
            crit_opt.zero_grad()
            crit_loss = (
                -(discriminator(real_images, conditions).mean()) +
                discriminator(fake_images, conditions).mean() +
                gradient_penalty(discriminator, real_images, fake_images, conditions, device, lambda_gp=LAMBDA)
            )
            crit_loss.backward()
            crit_opt.step()

        # Train Generator
        noise = get_noise(real_images.size(0), NOISE_DIM, device=device)
        fake_images = generator(noise, conditions)
        gen_opt.zero_grad()
        gen_loss = -discriminator(fake_images, conditions).mean()
        gen_loss.backward()
        gen_opt.step()

        # Record losses for each iteration
        generator_losses.append(gen_loss.item())
        critic_losses.append(crit_loss.item())

        # Accumulate losses for averaging at the epoch level
        epoch_gen_loss += gen_loss.item()
        epoch_crit_loss += crit_loss.item()

        # Increment iteration count
        iteration += 1

    # Calculate average losses for the epoch
    avg_gen_loss = epoch_gen_loss / len(dataloader)
    avg_crit_loss = epoch_crit_loss / len(dataloader)

    # Print average losses after each epoch
    print(f"Epoch {epoch+1}: Generator Loss: {avg_gen_loss:.4f}, Critic Loss: {avg_crit_loss:.4f}")

    # Display Real vs Fake Images Every 2 Epochs
    if (epoch + 1) % 20 == 0:
        noise = get_noise(6, NOISE_DIM, device=device)
        fake_samples = generator(noise, nn.functional.one_hot(labels[:6], CONDITION_DIM).float().to(device))
        show_real_and_fake_images(real_images[:6], fake_samples, title=f"Epoch {epoch+1}: Real vs Fake")

        # Plot losses against iterations
        plt.figure(figsize=(10, 5))
        plt.plot(generator_losses, label="Generator Loss")
        plt.plot(critic_losses, label="Critic Loss")
        plt.title(f"Losses per Iteration (up to Epoch {epoch+1})")
        plt.xlabel("Iterations")
        plt.ylabel("Loss")
        plt.legend()
        plt.show()

    # FID Calculation Every 10 Epochs
    if (epoch + 1) % 10 == 0:
        real_images_uint8 = denormalize_to_uint8(real_images[:BATCH_SIZE])
        fake_images_uint8 = denormalize_to_uint8(fake_images[:BATCH_SIZE])

        fid_metric.update(real_images_uint8, real=True)
        fid_metric.update(fake_images_uint8, real=False)

        fid_score = fid_metric.compute().item()
        fid_scores.append(fid_score)
        fid_metric.reset()

        # Log and Print Metrics
        print(f"Epoch {epoch+1}: Generator Loss: {avg_gen_loss:.4f}, Critic Loss: {avg_crit_loss:.4f}, FID Score: {fid_score:.4f}")

        with open(log_file, mode="a", newline="") as f:
            writer = csv.writer(f)
            writer.writerow([epoch + 1, avg_gen_loss, avg_crit_loss, fid_score])

    # Save Models Every 20 Epochs
    if (epoch + 1) % 20 == 0:
        torch.save(generator.state_dict(), f"models/generator_epoch_{epoch+1}.pth")
        torch.save(discriminator.state_dict(), f"models/discriminator_epoch_{epoch+1}.pth")
        print(f"Models saved at epoch {epoch + 1}")
```
